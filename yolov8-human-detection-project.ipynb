```python
!unzip human_detection_dataset.zip
```

```python
import yaml

dataset_info = {
    'train' : './train/images',
    'val' : './val/images',
    'path' : '../human_detection_dataset',
    'nc' : 1,
    'names' : ['Human']
}
with open('human_detection_dataset/data.yaml', 'w+') as f:
    doc = yaml.dump(dataset_info, f, default_flow_style = None, sort_keys = False)
```

# 4. Training model

```python
import wandb
import os
os.environ["WANDB_API_KEY"] = ""

wandb.init(project='human detection project')
```

```python
%cd ultralytics
!yolo train model=yolov8s.pt data=../human_detection_dataset/data.yaml epochs=20 imgsz=640
```

```python
wandb.finish()
```
